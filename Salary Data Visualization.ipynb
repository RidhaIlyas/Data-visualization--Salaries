```python
data.drop(columns='salary',inplace=True)
```

```python
data.groupby(['experience_level','job_title'])['salary_in_usd'].max().to_frame()
```

```python
data.sort_values(by='salary_in_usd',ascending=False)
```

# top paid jobs

```python
df=data.sort_values(by='salary_in_usd',ascending=False).head(10)
plt.bar(df['job_title'],df['salary_in_usd'])
plt.xticks(rotation=90);
```

* the plot shows that among the top 7 paid jobs data scientists are highly paid and the head of data science are the least paid

# effect of company size on salary

```python
plt.barh(data['company_size'],data['salary_in_usd'])
```

* large companies pay comparitively more salary followed by small companies 
* medium companies are the ones paying less salary when compared to large and small companies

# Location Distribution

```python
data['company_location'].value_counts().plot(kind='line',marker='*')
plt.xticks(rotation=90);
plt.title('location distribution of companies')
plt.ylabel('no.of companies');
```

* around 3000 companies are located in US whereas PT,PL,HR,ID,MD,MA and CR has almost same level of company distributions

# How remote ratio has changed with year and effect of company size on this

```python
sns.lineplot(x='work_year',y='remote_ratio',data=data,marker='o',hue='company_size');
```

* small companies has an increased shift in remote ratio
* in medium companies, though they showed an increase in 2021, sudden decline is noted towards 2022 and 2023
* in large companies,the ratio remained a little stable from 2020 to 2022, however a sudden decline in the ratio was noted from 2022 to 2023

# Trends in employment type

```python
sns.barplot(x='work_year',y='employment_type',data=data);
```

* the employment type has remained stable throughout all the years

# Distribution of Salary currency in the data

```python
data['salary_currency'].value_counts().plot(kind='barh')
```

* salary currency are mainly given in US dollars followed by Euros and Great Britain Pounds

# Salary Distribution in various locations

```python
data.groupby(['company_location'])['salary_in_usd'].max().head(8).sort_values(ascending=False).plot(kind='pie',autopct='%1.1f%%')
```

* salary distribution is equal in the area AE and BA with 18.1% with the least distribution in the area AL

# Relationship between Experience level and salary

```python
sns.lineplot(x='experience_level',y='salary_in_usd',data=data)
```

* Highest salary is plotted for Executive level followed by Senior Executives and Mid level, the least of all salaries are plotted for Entry level
